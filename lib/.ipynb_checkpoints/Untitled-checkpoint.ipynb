Our interest here is to inform users the “best” loan to invest. But how to define the "best" loan? 
Typically, higher returns come with higher risks. But inverstors should all agree that they certainly don't wan t to spend theit money with no return at all, in other words, they don't want to pick notes that will default in future and lose all of money. 
So, to simplfy the problem, I decided to limit my analysis scope to a classification problem by answering how to use  labels/attributes to identify a "good" note, and therefore assist inversors' investment decisions.

### Handling missing values
    - If the number of missing values is small, then simply drop those rows during training 
    - If the number of missing values is substatial, consider two situations: 
       1. randomly missing 
       2. systematically missing 

```python
def miss_value_table(dt):
    miss_val_cnt = dt.isnull().sum()
    miss_val_pct = dt.isnull().sum()/len(dt)*100
    miss_val_table  = pd.concat([miss_val_cnt,miss_val_pct],axis =1 )
    miss_val_table.columns = ['Number of missing values', "% of missing values"]
    miss_val_table_sec = miss_val_table.loc[miss_val_table['Number of missing values']!=0, : ].sort_values(by=['Number of missing values'], ascending = False)
    return miss_val_table_sec
```

```python
app_type_count  = loan_df.application_type.value_counts() 
label = ['Individual', 'Joint App']
plt.pie(app_type_count, explode=(0,0), labels = label, shadow = True, autopct='%1.1f%%', startangle = 90)
plt.title("Distribution of Loan Type")
plt.show()
# less than 3% of loans are of joint type 
```

```python
# default rate vs loan type 
c = loan_df.groupby(['application_type', 'default'])['default'].count()
default_by_app_type = c/c.groupby(level=0).sum()
default_by_app_type= default_by_app_type.apply(lambda x:100 * x).rename('percentage').reset_index()
```

```python
indices = np.array([0,1])
width = 0.3
names = ['Individual','Joint App' ]
bars1 = default_by_app_type[default_by_app_type.default == 'No'].percentage.tolist()
bars2 = default_by_app_type[default_by_app_type.default == 'Yes'].percentage.tolist()

plt.figure(figsize=(5,4))
plt.bar(indices-width,bars1,width,color='lightcoral', label = 'Not Default')
plt.bar(indices,bars2,width,color='cadetblue', label = 'Default')
plt.xticks(indices, names)
plt.ylabel('Percentage')
plt.title("Default Rate, Breaking Down by Application Type", fontsize = 10)
plt.legend(bbox_to_anchor=(1, 1), loc='upper left', ncol=1, fontsize = 8)
plt.show()
```

- Given the default rates of two types of application are almost the same, which indicates that the presence of co-borrowers don't help to lower the default risk; so the application type has alomost no prediction power in identifying good or bad loan 
- Plus, with small amount of joint application loans, there is little incentive to choose joint over individual.

### **Months since event variables**
- check if it is systematic missing, that is missing because not applicable
- check the summary statistics and prediction power of this feature
- impute missing values

```python
# mths_since_last_record: measure how long it has been since last derogatory public record 
# This feature is systematic missing value - it's missing when the borrower doesn't have any derogatory public record 
# (derogatory public records usually consist of bankruptcy filings, civil court judgments, foreclosures and tax liens)
g = indivapp_df.mths_since_last_record.isnull().groupby([indivapp_df['pub_rec']]).sum().astype(int).reset_index(name = 'countna')
g.loc[g['countna'] > 0,:]
```

```python
# mths_since_last_record range from 0 (record occurred less than 1 month) to 121(10 years)
print( indivapp_df.mths_since_last_record.describe() ) 
plt.figure(figsize=(4,5))
plt.boxplot(indivapp_df.mths_since_last_record.dropna())
plt.show()
```

```python
# check how this variable associate 
bins = [0, 12, 36, 60, 96, 125] #np.arange(1000,15000,1000).tolist()  #
labels = ['<1 year', '1-3', '3-5' ,'5-8', '>8']
indivapp_df['derog_grouped'] = pd.cut(indivapp_df['mths_since_last_record'].dropna(), bins, right=0, labels=labels)
```

```python
gb = indivapp_df.groupby(['derog_grouped','default'])['default'].count()
gb/gb.groupby(level=0).sum()
```

```python
gb_null = indivapp_df.mths_since_last_record.isnull().groupby(indivapp_df.default).count()
gb_null/gb_null.sum()
```

- With low values, say less than a year, the default rate is around 30%, as the number of months since last derogatory record inceases, the default rate is round 25%; 
- The population without derogatory record has default rate around 24%. 
- So there is definitely some predict power in it
- Since the problem we are trying to solve here is classification, and the real value here is equal to or above 0, therefore encode missing value with negative number will split this population.
